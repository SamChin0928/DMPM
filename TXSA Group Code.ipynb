## Import NLTK packages, import text corpus into Python and Tokenizing the words in text corpus.

```python
import nltk
nltk.download('punkt')
import pandas as pd
import re
from nltk.tokenize import word_tokenize

print(nltk.__version__,"\n")

tokens = []
sentence_list = []

# Importing Text Corpus
f = open("/content/drive/MyDrive/TXSA Data Folder/Group Assignment Data/Text Corpus.txt", "r")
text = f.read()
f.close()
print("Sentences in Text Corpus:")
print(text,"\n")

# Split the Text Corpus based on lines
text = text.splitlines()
print("List of Sentences:")
print(text,"\n")

# Removing Padding 
for sentence in text:
  sentence_list.append(re.sub("<s> | </s>","",sentence))
print("Sentences after padding as been removed:")
print(sentence_list,"\n")

# Tokenize the sentences and pre-process the tokenized sentences
for x in range(len(sentence_list)):
  temp = list(map(str.lower, nltk.tokenize.word_tokenize(sentence_list[x])))
  tokens.append(temp)

print("List of tokens:")
print(tokens,"\n")
```

## Unsmoothed Unigram Language Model

```python
from nltk.lm import MLE
from nltk.util import pad_sequence
from nltk.lm.preprocessing import padded_everygram_pipeline
n = 1
padded = []

# Re-add padding to the front and end of each sentence
for items in tokens:
  temp = list(pad_sequence(items, pad_left=True, left_pad_symbol="<s>", pad_right=True, right_pad_symbol="</s>", n=2))
  padded.append(temp)

train_data, padded_sents = padded_everygram_pipeline(n,padded)

unsmooth_unigram_model = MLE(n)
unsmooth_unigram_model.fit(train_data, padded_sents)

print("N-Gram in the Unsmoothed Unigram Model: ",unsmooth_unigram_model.counts)
print("Vocabulary in Unsmoothed Unigram Model: ",unsmooth_unigram_model.vocab)
print("\nProbabilty for Unsmoothed Unigram")
print("Model Score for '<s>'\t\t:", unsmooth_unigram_model.score('<s>'))
print("Model Score for '</s>'\t\t:",unsmooth_unigram_model.score('</s>'))
print("Model Score for 'He'\t\t:",unsmooth_unigram_model.score('he')) 
print("Model Score for 'read'\t\t:",unsmooth_unigram_model.score('read'))
print("Model Score for 'a'\t\t:",unsmooth_unigram_model.score('a'))
print("Model Score for 'book'\t\t:",unsmooth_unigram_model.score('book')) 
print("Model Score for 'I'\t\t:",unsmooth_unigram_model.score('i'))
print("Model Score for 'different'\t:",unsmooth_unigram_model.score('different'))
print("Model Score for 'my'\t\t:",unsmooth_unigram_model.score('my')) 
print("Model Score for 'Danielle'\t:",unsmooth_unigram_model.score('danielle')) 
```

## Prediction of sentiments for each review found in the data set using the Rule-Based Unsupervised Technique

```python
import dataclasses
import nltk
nltk.download('vader_lexicon')
import pandas as pd 
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

data = pd.read_csv("/content/drive/MyDrive/TXSA Data Folder/Group Assignment Data/Musical_Instruments_Reviews.csv") 

reviews = data['Reviews']
print("\nTotal reviews in the Musical Instrument Review dataset is",reviews.count())
print()
sentiments = []

for line in reviews:
    sent = sia.polarity_scores(line)
    if (sent['compound'] == 0):
        sentiments.append('Neutral')
    else:
        if (sent['compound'] > -0):
            sentiments.append('Positive')
        else:
            sentiments.append('Negative')

data['Sentiments'] = sentiments
print(data[:10])

# Export compund data frame into csv file format
data.to_csv ('/content/test.csv', index = False, header=True)
```

## Supervised sentiment classification model using Naïve Bayes Classifier from NLTK


```python
from nltk.metrics.aline import R
import pandas as pd 

data = pd.read_csv('/content/test.csv')


print(len(data))

#Separate data into positive negative and neutral set

positive_sent = data.loc[data['Sentiments'] == 'Positive']
negative_sent = data.loc[data['Sentiments'] == 'Negative']
neutral_sent =  data.loc[data['Sentiments'] == 'Neutral']

print("Length of Positive Labeled Data",len(positive_sent))
print("Length of Negative Labeled Data",len(negative_sent))
print("Length of Neutral Labeled Data",len(neutral_sent))

print(positive_sent)
print(negative_sent)
print(neutral_sent)



#Tokenization for Data
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')

untreated_positive_tokens = [word_tokenize(words.lower()) for words in positive_sent['Reviews']]
untreated_negative_tokens = [word_tokenize(words.lower()) for words in negative_sent['Reviews']]
untreated_neutral_tokens = [word_tokenize(words.lower()) for words in neutral_sent['Reviews']]

print(untreated_positive_tokens[:10])
print(untreated_negative_tokens[:10])
print(untreated_neutral_tokens[:10])


#Remove stopwords and punctuation
nltk.download('stopwords')
from nltk.corpus import stopwords
import string

stoptokens = stopwords.words("english") + list(string.punctuation)

filtered_positive_tokens = []
filtered_negative_tokens = []
filtered_neutral_tokens = []

for item in untreated_positive_tokens:
  if item not in stoptokens:
    filtered_positive_tokens.append(item)

for item in untreated_negative_tokens:
  if item not in stoptokens:
      filtered_negative_tokens.append(item)

for item in untreated_neutral_tokens:
  if item not in stoptokens:
      filtered_neutral_tokens.append(item)

print(filtered_positive_tokens[:100])
print(filtered_negative_tokens[:100])
print(filtered_neutral_tokens[:100])

#Lemmatization
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
from nltk.tag import pos_tag


def lemmatize_tokens(token):
  lemma_list = []
  lemma_tokens = []
  for item in token:
    for elem, tag in pos_tag(item):
        if tag.startswith('NN'):
          pos = 'n'
        elif tag.startswith('VB'):
          pos = 'v'
        else:
          pos = 'a'
        lemma = lemmatizer.lemmatize(elem,pos)
        lemma_list.append(lemma)
    lemma_tokens.append(lemma_list)
    lemma_list = []
  return lemma_tokens

lemma_positive_tokens = lemmatize_tokens(filtered_positive_tokens)
lemma_negative_tokens = lemmatize_tokens(filtered_negative_tokens)
lemma_neutral_tokens = lemmatize_tokens(filtered_neutral_tokens)

print("Lemmatized Positive Tokens:", lemma_positive_tokens[:100])
print("Lemmatized Negative Tokens:", lemma_negative_tokens[:100])
print("Lemmatized Neutral Tokens:", lemma_neutral_tokens[:100])


#Feature Extraction : Converting data into dictionary data structure

def sentiment_features(data):
  for elem in data:
    yield dict([word,True] for word in elem)

positive_token_set = sentiment_features(lemma_positive_tokens)
negative_token_set = sentiment_features(lemma_negative_tokens)
neutral_token_set = sentiment_features(lemma_neutral_tokens)

positive_set = [(word, "Positive") for word in positive_token_set]
negative_set = [(word, "Negative") for word in negative_token_set]
neutral_set = [(word, "Neutral") for word in neutral_token_set]

dataset = positive_set + negative_set +neutral_set

#Spliitting the dataset
from sklearn.model_selection import train_test_split

train_set,test_set = train_test_split(dataset,test_size = 0.3,random_state =456)
print("Length of Training Set:",len(train_set))
print("Length of Test Set:",len(test_set))

#Naive Bayes Model
from nltk import classify
import nltk 

classifier = nltk.NaiveBayesClassifier.train(train_set)
classifier.show_most_informative_features(25)

#Evaluation Metrics
print("Accuracy: {:4.2f}".format(classify.accuracy(classifier, test_set)))

import collections

refsets = collections.defaultdict(set)
testsets = collections.defaultdict(set)

for i, (feats, label) in enumerate(test_set):
    refsets[label].add(i)
    observed = classifier.classify(feats)
    testsets[observed].add(i)

print('Positive Precision:{}'.format(nltk.precision(refsets['Positive'], testsets['Positive']))) 
print('Negative Precision:{}'.format(nltk.precision(refsets['Negative'], testsets['Negative']))) 
print('Neutral Precision:{}'.format(nltk.precision(refsets['Neutral'], testsets['Neutral']))) 
print('Positive Recall:{}'.format(nltk.recall(refsets['Positive'], testsets['Positive']))) 
print('Negative Recall:{}'.format(nltk.recall(refsets['Negative'], testsets['Negative']))) 
print('Neutral Recall:{}'.format(nltk.recall(refsets['Neutral'], testsets['Neutral']))) 
print('Positive F-1 Score:{}'.format(nltk.f_measure(refsets['Positive'], testsets['Positive']))) 
print('Negative F-1 Score:{}'.format(nltk.f_measure(refsets['Negative'], testsets['Negative']))) 
print('Neutral F-1 Score:{}'.format(nltk.f_measure(refsets['Neutral'], testsets['Neutral']))) 

#Confusion Matrix
true_feature = [feature for (feature, label) in test_set]
true_label = [label for (feature, label) in test_set]
true_label_predicted =  [classifier.classify(f) for f in true_feature] 

confusionmatrix = nltk.ConfusionMatrix(true_label,true_label_predicted)
print(confusionmatrix.pretty_format(sort_by_count=True))
```
